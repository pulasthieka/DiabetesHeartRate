```python
import pandas as pd
dataset = pd.read_csv('swell_training_data.csv')
dataset.head()
```

```python
dataset.columns
```

```python
def labelDataset(row):
#     label the dataset 1- healthy , 0- diabetic
    if row['label'] =="healthy":
        return  1
    else:
        return 0
dataset['label'] = dataset.apply (lambda row: labelDataset(row), axis=1)
train_labels = dataset['label']
train_dataset = dataset.drop(['label'],axis=1)
train_dataset.head()
```

```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
model = keras.Sequential([
    keras.layers.InputLayer(input_shape=(9)),
    keras.layers.Dense(20, activation='sigmoid'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.summary()
```

```python
train_stats = train_dataset.describe()
train_stats = train_stats.transpose()
print(train_stats)
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
# normed_test_data = norm(test_dataset)
```

```python
model.fit(normed_train_data, train_labels, epochs=20)
```

```python
test_dataset = pd.read_csv('swell_testing_data.csv')
# test_data = test_dataset.drop([ 'SDSD', 'SDRR_RMSSD', 
#        'pNN25', 'SD1', 'SD2', 'KURT', 'SKEW', 'MEAN_REL_RR',
#        'MEDIAN_REL_RR', 'SDRR_REL_RR', 'RMSSD_REL_RR', 'SDSD_REL_RR',
#        'SDRR_RMSSD_REL_RR', 'KURT_REL_RR', 'SKEW_REL_RR', 'VLF', 'VLF_PCT',
#        'LF_PCT', 'LF_NU', 'HF_PCT', 'HF_NU', 'TP', 'LF_HF',
#        'sampen', 'higuci', 'datasetId', 'condition'], axis=1)

def labelDataset(row):
#     label the dataset 1- diabetic , 0- non-diabetic
    if row['label'] =="healthy":
        return  1
    else:
        return 0
test_dataset['label'] = test_dataset.apply (lambda row: labelDataset(row), axis=1)
test_labels = test_dataset['label']
test_data = test_dataset.drop(['label'],axis =1)
normed_test_data = norm(test_data)

test_loss, test_acc = model.evaluate(normed_test_data,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

```python
model.save("diabetesHRV.h5")
print("Saved model to disk")
```
